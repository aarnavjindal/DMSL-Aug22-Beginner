```python
students = [
    { "name": "A", "marks": 50 },
    { "name": "B", "marks": 100 },
    { "name": "C", "marks": 40 },
    { "name": "D", "marks": 80 },
    { "name": "E", "marks": 70 },
]
```

```python
# Throws error
sorted_students = sorted(students)
```

```python
sorted_students = sorted(
    students,
    key = lambda x : x["marks"]
)

print(sorted_students)
```

```python
# Higher Order Functions

def gen_exp(n) :
  
  def exp(x) :
    return x ** n

  return exp
```

```python
exp_3 = gen_exp(3)
print(exp_3(2))
print(exp_3(4))

exp_5 = gen_exp(5)
print(exp_5(3))
```

```python
def gen_exp(n) : 
  
  def exp(x) :
    a = (x ** n)
  
    def exp1(y) : 
      return a//y 
    
    return exp1
  
  return exp

gen_exp(2)(3)(2)
```

```python
exp_5 = gen_exp(5)(3)
print(type(exp_5), exp_5)

exp_5 = gen_exp(5)
print(type(exp_5), exp_5(3))
```
